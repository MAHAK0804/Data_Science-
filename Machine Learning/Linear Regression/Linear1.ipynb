```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('C:\\Users\\Hp\\Desktop\\Data_science-main\\Datasets\\homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area(sqr ft)')
plt.ylabel('Prices(US $)')
plt.title('home prices')
plt.scatter(df.area,df.prices,color="green",marker="+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.prices)
```

```python
reg.predict(np.array([[2600]]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Area',fontsize = 20)
plt.ylabel('Prices',fontsize = 20)
plt.title('Home Prices',fontsize = 20)
plt.scatter(df[['area']],df.prices,color = 'green',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color = 'blue')
```

```python
d = pd.read_csv('C:\\Users\\Hp\\Desktop\\Data_science-main\\Datasets\\area.csv')
d.head(3)
```

```python
p= reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_csv('prediction.csv',index=False)
```

```python
ds = pd.read_csv('C:\\Users\\Hp\\Desktop\\Data_science-main\\Datasets\\canada_per_capita_income.csv')
ds.head(3)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.title('Net income')
plt.scatter(ds[['year']],ds.per_capita_income,color= 'green',marker='+')
```

```python
reg.predict(np.array([[2020]]))
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.title('Net income')
plt.scatter(ds[['year']],ds.per_capita_income,color= 'green',marker='+')
plt.plot(ds.year,reg.predict(ds[['year']]),color='blue')
```
